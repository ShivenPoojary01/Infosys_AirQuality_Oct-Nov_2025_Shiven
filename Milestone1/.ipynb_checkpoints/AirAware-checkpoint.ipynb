```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

df[df.select_dtypes(include='number').columns] = num_imputer.fit_transform(df.select_dtypes(include='number'))
df[df.select_dtypes(include='object').columns] = cat_imputer.fit_transform(df.select_dtypes(include='object'))

```

```python
df.drop_duplicates(inplace=True)
```

```python
for col in df.select_dtypes(include='object'):
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Convert the sampling_date column to datetime
df['sampling_date'] = pd.to_datetime(df['sampling_date'], errors='coerce')

# Extract the year
df['Year'] = df['sampling_date'].dt.year

# Calculate average pollution levels per year
pollution_by_year = df.groupby('Year')[['so2', 'no2', 'rspm', 'spm', 'pm2_5']].mean().reset_index()

# Plot timeline for all pollutants
plt.figure(figsize=(12, 6))
for col in ['so2', 'no2', 'rspm', 'spm', 'pm2_5']:
    if col in pollution_by_year.columns:
        sns.lineplot(data=pollution_by_year, x='Year', y=col, marker='o', label=col.upper())

plt.title('Average Air Pollution Levels Over the Years', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Average Concentration (g/m)')
plt.grid(True)
plt.legend(title='Pollutant')
plt.show()
```
